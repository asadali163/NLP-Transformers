```python
from transformers import pipeline, set_seed

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import textwrap
from pprint import pprint
```

```python
# Let's see inside our text file
!cat robert_frost.txt
```

As it has empty lines, so we will take into account all of these

```python
lines = [line.strip() for line in open('robert_frost.txt')]
lines = [line for line in lines if len(line) > 0]
```

```python
# Load the model
gen = pipeline('text-generation')
```

So the GPT 2 model is loaded by default

```python
len(lines)
```

```python
set_seed(1234)
```

```python
lines[350]
```

```python
gen(lines[350])
```

```python
pprint(_)
```

```python
pprint(gen(lines[350], max_length=20))
```

```python
pprint(gen(lines[350], max_length=20, num_return_sequences=3))
```

```python
# Define wrap function
def wrap(x):
    return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

Let's do some kind of poetry stuff with text generation model<br>
we will use the first line as input to model, then use even one later

```python
out = gen(lines[0], max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prev = "Two roads diverged in a yellow wood, but the bridge just wasn't visible, meaning the bridge took over the road on the left."
```

```python
out = gen(prev + '\n' + lines[2], max_length=60)
print(wrap(out[0]['generated_text']))
```

```python
prev = """Two roads diverged in a yellow wood, but the bridge just wasn't
visible, meaning the bridge took over the road on the left.
And be one traveler, long I stood, the sound of footsteps and cars behind me.  A
low chime of horn."""
```

```python
out = gen(prev + '\n' + lines[4], max_length=90)
print(wrap(out[0]['generated_text']))
```

As it looks like a poetry, but not exactly the poetry. For that we need to fine tune it first

As GPT-2 is not trained on poetry that is why it is not good at. <br>
Let's look at general text and see how it perform

```python
prompt = "Strong foundation in mathematics is necessary for someone who want to learn deep learning"
out = gen(prompt, max_length=300)
print(wrap(out[0]['generated_text']))
#
```
