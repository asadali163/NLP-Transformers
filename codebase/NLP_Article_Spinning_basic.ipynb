```python
#### Let's write a function that automatically mask and replace words in whole document
def replace_word_with_masked_model(text):
    words = text.split()
    new_text = []

    for i, word in enumerate(words):
        # Create a copy of the words list with the current word masked
        masked_words = words[:]
        masked_words[i] = mlm.tokenizer.mask_token  # [MASK]

        # Convert words to a single string
        masked_text = ' '.join(masked_words)

        # Predict the masked word
        results = mlm(masked_text)

        # Get the most likely predicted word
        predicted_word = results[0]['token_str']

        # Replace the masked word with the predicted word
        new_text.append(predicted_word)

    # Join the list of words to form the new text
    return ' '.join(new_text)
```
